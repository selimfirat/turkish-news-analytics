```python
from data import Data 
from sklearn.metrics import classification_report
from sklearn.neighbors.nearest_centroid import NearestCentroid
from sklearn.model_selection import ShuffleSplit
from plot_learning_curve import plot_learning_curve
import matplotlib.pyplot as plt

%matplotlib inline

data = Data()

def classify(train_x, train_y, test_x, test_y):
    
    cv = ShuffleSplit(n_splits=10, test_size=0.2, random_state=0)
    est = NearestCentroid()

    plot_learning_curve(est, "Nearest Centroid", train_x, train_y, cv=cv)
    plt.show()
    
    est = est.fit(X=train_x, y = train_y)
    pred_y = est.predict(test_x)
    print(classification_report(pred_y, test_y))
```

```python
train_x, train_y, test_x, test_y = data.count_vectorized()
classify(train_x, train_y, test_x, test_y)
```

```python
train_x, train_y, test_x, test_y = data.tfidf_vectorized()
classify(train_x, train_y, test_x, test_y)
```
