```python
from data import Data 
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import ShuffleSplit
from plot_learning_curve import plot_learning_curve
import matplotlib.pyplot as plt
import json

%matplotlib inline

def classify(train_x, train_y, test_x, test_y, classifier_params):
    
    cv = ShuffleSplit(n_splits=10, test_size=0.2, random_state=0)
    logreg = LogisticRegression(**classifier_params)

    #plot_learning_curve(logreg, "Logistic Regression", train_x, train_y, cv=cv)
    #plt.show()
    
    logreg = logreg.fit(X=train_x, y = train_y)
    pred_y = logreg.predict(test_x)
    print(classification_report(pred_y, test_y))
    
```
